```python
#Question 1
#Introduction:
#Special thanks to: https://github.com/justmarkham for sharing the dataset and materials. Occupations
#Step 1. Import the necessary libraries
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called users
#Step 4. Discover what is the mean age per occupation
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
#Step 6. For each occupation, calculate the minimum and maximum ages
#Step 7. For each combination of occupation and sex, calculate the mean 
```

```python
#Step 1. Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Import the dataset
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

#Step 3. Assign it to a variable called users
users = pd.read_csv(url ,sep='|')
print(users)

# Step 4: Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
# Count the number of males per occupation
males_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count()

# Count the total number of individuals per occupation
total_per_occupation = users.groupby('occupation')['gender'].count()

# Calculate the male ratio per occupation
male_ratio_per_occupation = males_per_occupation / total_per_occupation

# Sort the male ratios
male_ratio_sorted = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: For each occupation, calculate the minimum and maximum ages
min_age_per_occupation = users.groupby('occupation')['age'].min()
max_age_per_occupation = users.groupby('occupation')['age'].max()

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_per_occupation_and_gender = users.groupby(['occupation', 'gender'])['age'].mean()

# Print the results
print("Mean age per occupation:\n", mean_age_per_occupation)
print("\nMale ratio per occupation:\n", male_ratio_sorted)
print("\nMinimum age per occupation:\n", min_age_per_occupation)
print("\nMaximum age per occupation:\n", max_age_per_occupation)
print("\nMean age per occupation and gender:\n", mean_age_per_occupation_and_gender)
```

```python
import pandas as pd
import numpy as np

# Step 2: Importing the dataset
file_path = 'wind.txt'

# Step 3: Reading the data and replacing the first 3 columns with a proper datetime index
data = pd.read_csv(file_path, sep="\s+", parse_dates=[[0, 1, 2]])

# Step 4: Fixing the year issue in the date column (from 2061 to 1961)
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(lambda d: d.replace(year=d.year - 100) if d.year > 1979 else d)

# Step 5: Setting the right dates as the index
data.set_index('Yr_Mo_Dy', inplace=True)
data.index = pd.to_datetime(data.index)

# Step 6: Compute how many values are missing for each location over the entire record
missing_values = data.isnull().sum()

# Step 7: Compute how many non-missing values there are in total
non_missing_values = data.count().sum()

# Step 8: Calculate the mean windspeeds over all the locations and all the times
mean_windspeed = data.mean().mean()

# Step 9: Creating a DataFrame called loc_stats with min, max, mean windspeeds and standard deviations at each location
loc_stats = data.describe().loc[['min', 'max', 'mean', 'std']]

# Step 10: Creating a DataFrame called day_stats with min, max, mean windspeed, and standard deviations across all locations for each day
day_stats = data.agg(['min', 'max', 'mean', 'std'], axis=1)

# Step 11: Find the average windspeed in January for each location
january_avg = data[data.index.month == 1].mean()

# Step 12: Downsample the record to a yearly frequency for each location
yearly_avg = data.resample('A').mean()

# Step 13: Downsample the record to a monthly frequency for each location
monthly_avg = data.resample('M').mean()

# Step 14: Downsample the record to a weekly frequency for each location
weekly_avg = data.resample('W').mean()

# Step 15: Calculate the min, max, mean windspeeds, and standard deviations for the first 52 weeks
# Assuming the first week starts on January 2, 1961
start_date = "1961-01-02"
end_date = "1961-12-31"
data_52_weeks = data[start_date:end_date]
weekly_stats_first_year = data_52_weeks.resample('W').agg(['min', 'max', 'mean', 'std'])

# Printing the results from the analysis

print("Wind Statistics Analysis:")
print("=====================================")

# Step 6: Missing Values for Each Location
print("Step 6: Missing Values for Each Location")
print(missing_values)

# Step 7: Total Non-Missing Values
print("\nStep 7: Total Non-Missing Values")
print(f"Total non-missing values in the dataset: {non_missing_values}")

# Step 8: Mean Windspeeds over All Locations and Times
print("\nStep 8: Mean Windspeeds over All Locations and Times")
print(f"Mean windspeed across all locations and times: {mean_windspeed:.2f} knots")

# Step 9: Statistics at Each Location (loc_stats)
print("\nStep 9: Statistics at Each Location (loc_stats)")
print(loc_stats)

# Step 10: Daily Statistics across All Locations (day_stats)
print("\nStep 10: Daily Statistics across All Locations (day_stats)")
print(day_stats.head())  # Displaying only the first few rows for brevity

# Step 11: Average Windspeed in January for Each Location
print("\nStep 11: Average Windspeed in January for Each Location")
print(january_avg)

# Step 12: Yearly Average Windspeed for Each Location
print("\nStep 12: Yearly Average Windspeed for Each Location")
print(yearly_avg.head())  # Displaying only the first few years for brevity

# Step 13: Monthly Average Windspeed for Each Location
print("\nStep 13: Monthly Average Windspeed for Each Location")
print(monthly_avg.head())  # Displaying only the first few months for brevity

# Step 14: Weekly Average Windspeed for Each Location
print("\nStep 14: Weekly Average Windspeed for Each Location")
print(weekly_avg.head())  # Displaying only the first few weeks for brevity

# Step 15: Weekly Statistics for the First 52 Weeks
print("\nStep 15: Weekly Statistics for the First 52 Weeks")
print(weekly_stats_first_year.head())  # Displaying only the first few weeks for brevity


```

```python
#Step 1. Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
# Assuming the file is a tab-separated values file, we'll use read_csv with sep='\t'
file_path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

#Step 3. Assign it to a variable called chipo.
chipo = pd.read_csv(file_path, sep='\t')

# Step 4: See the first 10 entries
first_10_entries = chipo.head(10)

# Step 5: Number of observations in the dataset
num_observations = chipo.shape[0]

# Step 6: Number of columns in the dataset
num_columns = chipo.shape[1]

# Step 7: Print the name of all the columns
column_names = chipo.columns.tolist()
print(column_names)

# Step 8: How is the dataset indexed
dataset_index = chipo.index

# Step 9: The most-ordered item
most_ordered_item = chipo.groupby('item_name').quantity.sum().idxmax()

# Step 10: For the most-ordered item, how many items were ordered
most_ordered_item_quantity = chipo.groupby('item_name').quantity.sum().max()

# Step 11: The most ordered item in the choice_description column
most_ordered_choice = chipo.groupby('choice_description').quantity.sum().idxmax()

# Step 12: How many items were ordered in total
total_items_ordered = chipo.quantity.sum()

# Step 13: Turn the item price into a float and check the item price type
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)
item_price_type = chipo['item_price'].dtype

# Step 14: Revenue for the period in the dataset
total_revenue = (chipo['quantity'] * chipo['item_price']).sum()

# Step 15: How many orders were made in the period
total_orders = chipo.order_id.nunique()

# Step 16: Average revenue amount per order
average_revenue_per_order = total_revenue / total_orders

# Print the results
first_10_entries, num_observations, num_columns, column_names, dataset_index, most_ordered_item, most_ordered_item_quantity, most_ordered_choice, total_items_ordered, item_price_type, total_revenue, total_orders, average_revenue_per_order
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Step 2: Importing the dataset
file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Creating a line plot
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per Capita', color='blue')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per Capita', color='red')

# Labeling the plot
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Count per 1000 People')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Step 2: Importing the dataset
file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Selecting data for the years 1900, 1950, and 2000
selected_years = data[data['Year'].isin([1900, 1950, 2000])]

# Creating a vertical bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(selected_years['Year']))

# Plotting marriages and divorces
bar1 = plt.bar(index, selected_years['Marriages_per_1000'], bar_width, label='Marriages per Capita', color='blue')
bar2 = plt.bar(index + bar_width, selected_years['Divorces_per_1000'], bar_width, label='Divorces per Capita', color='red')

# Labeling the plot
plt.xlabel('Year')
plt.ylabel('Count per 1000 People')
plt.title('Number of Marriages and Divorces per Capita in the U.S. for Years 1900, 1950, and 2000')
plt.xticks(index + bar_width / 2, selected_years['Year'])
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Step 2: Importing the dataset
file_path = 'roman-emperor-reigns.csv'
emperor_data = pd.read_csv(file_path)

# Counting the fraction of emperors that were assassinated
assassination_counts = emperor_data['Cause_of_Death'].value_counts()

# Creating a pie chart
plt.figure(figsize=(8, 8))
plt.pie(assassination_counts, labels=assassination_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Ensuring the pie chart is a circle

# Adding a title
plt.title('Fraction of Roman Emperors Assassinated')

# Show the plot
plt.show()
```
